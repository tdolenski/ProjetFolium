```python
import pandas as pd

d=r.json()

data = pd.json_normalize(d['results'])
data.isna().sum()
```

```python
#On nettoie les colonnes
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
#On supprime des colonnes

#del data['sorted_tags.transport']
#del data['essentials']
#del data['sorted_tags.kitchen']
del data['sorted_tags.price']
del data['sorted_tags.essentials']
del data['sorted_tags.zeplus']
del data['address.updated_at']

data
```

```python
arrondissement='75011'
spe='vins'

data[(data['address.postal_code']==arrondissement) & (data['cravings'].astype(str).str.contains(spe))][['name','address.street_name_1','phone_num']]
```

```python
#On sépare latitude et longitude
long=[]
lat=[]

for elem in data['address.position']:
    lat.append(elem[0])
    long.append(elem[1]) 
    
data['address_position_lat']=lat
data['address_position_long']=long

```

```python
#ON AFFICHE LA CARTE
import folium 
from folium.plugins import MarkerCluster

rest_map = folium.Map(location=[48.85,2.37], zoom_start=13, tiles='OpenStreetMap')

marker_cluster = folium.plugins.MarkerCluster().add_to(rest_map)

arrondissement='75011'
spe='Cuisine du marché'

for i in data[(data['address.postal_code']==arrondissement) & (data['cravings'].astype(str).str.contains(spe))][['name','address_position_lat','address_position_long']].index:
    marker = folium.Marker(location=(round(data['address_position_lat'].loc[i],6),round(data['address_position_long'].loc[i],6)), popup=(str(data['name'].loc[i]))).add_to(marker_cluster)

rest_map

#rest_map.save('carte.html')
```
